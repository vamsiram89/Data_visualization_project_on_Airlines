```python
# ==========================================
#  IMPORT LIBRARIES
# ==========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ==========================================
#  LOAD DATASET & BASIC CLEANING
# ==========================================
df = pd.read_csv("/content/sample_data/airlines_flights_data.csv")
df
```

```python
# drop unnecessary empty columns:
df = df.drop(['Unnamed: 12','Unnamed: 13'],axis = 1,errors = 'ignore')
df
```

```python
# find the number of rows and columns
print("shape:",df.shape)
```

```python
# to get the information about the data
df.info()
```

```python
# to check the null values
df.isnull().sum()
```

```python
# ==========================================
# 3. EXPLORATION QUESTIONS
# ==========================================
# ---------- Q1. Cheapest airline overall ----------
# Calculate average price per airline


# Calculate average price by airline
cheapest_airlines = df.groupby('airline')['price'].mean().reset_index()

# Sort by price
cheapest_airlines = cheapest_airlines.sort_values('price')

# Print cheapest airline
print("Cheapest Airline:")
print(cheapest_airlines.iloc[0])

# Simple colorful barplot
plt.figure(figsize=(8,5))
sns.barplot(data=cheapest_airlines, x='airline', y='price', palette='viridis')

plt.title("Average Ticket Price by Airline")
plt.xticks(rotation=45)
plt.show()
```

```python

# ---------- Q2. Price vs days left ----------
# Draw the How does ticket price vary with number of days left before departure?


# Calculate average price per days_left
price_vs_days = df.groupby('days_left')['price'].mean().reset_index()

# Set seaborn style
sns.set_style("whitegrid")
plt.figure(figsize=(8,5))

# Line plot with markers
sns.lineplot(data=price_vs_days,x='days_left',y='price',marker='o',linewidth=2,color='purple')

plt.title("Average Ticket Price vs Days Left", fontsize=14)
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.tight_layout()
plt.show()
```

```python
# ---------- Q3. Most expensive routes ----------


# Grouping
Expensive_routes = df.groupby(['source_city','destination_city'])['price'].mean()\
                     .sort_values(ascending=False).head(5)

# Convert Series to DataFrame
Expensive_routes = Expensive_routes.reset_index()

# Create a combined route column
Expensive_routes['route'] = Expensive_routes['source_city'] + " → " + Expensive_routes['destination_city']

# Set seaborn style
sns.set_style("whitegrid")

plt.figure(figsize=(8,6))

# Colorful barplot
sns.barplot(x='route',y='price',data=Expensive_routes,palette='cool')

plt.title("Top 5 Most Expensive Routes", fontsize=14)
plt.xlabel("Route")
plt.ylabel("Average Price")
plt.xticks(rotation=45)

plt.show()
```

```python
# ---------- Q4. Economy vs Business ----------
#Does departure time affect ticket price?
#2.Average Price: Economy vs. Business Class

Airlines_avg = df.groupby('airline')['price'].mean()

print( "all airlines averages individully")
ALs_avg_indv = Airlines_avg.values

label = Airlines_avg.index
plt.pie(ALs_avg_indv, labels= label, autopct = "%1.1f%%")
plt.title("all airlines averages individully")
plt.show()

print("Average Price: Economy vs. Business Class " )

Economy_vs_business = df.groupby('class')['price'].mean()
values= Economy_vs_business.values
label= Economy_vs_business.index
plt.pie(values,labels=label,autopct='%1.1f%%')
plt.title("Average Price: Economy vs. Business Class")
plt.show()
```

```python
# ---------- Q5. Impact of departure time ----------

# Group and reset index
IODT = df.groupby('departure_time')['price'].mean().reset_index()

print("\nDeparture Time Impact:\n", IODT)

# Set seaborn theme
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

# Create colorful barplot
sns.barplot(x='departure_time',y='price',data=IODT,palette='magma')

plt.title("Average Price by Departure Time", fontsize=16, fontweight='bold')
plt.xlabel("Departure Time")
plt.ylabel("Average Ticket Price")

plt.xticks(rotation=45)

plt.show()
```

```python
# ---------- Q6. Airline with highest variability ----------
# Which airline has the highest variation (fluctuation) in ticket prices?
import seaborn as sns
import matplotlib.pyplot as plt
# Calculate standard deviation (price variability)
airline_high_var = df.groupby('airline')['price'].std().reset_index()
# Sort descending
airline_high_var = airline_high_var.sort_values(by='price', ascending=False)
print("\nPrice Variability:\n", airline_high_var)

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(data=airline_high_var,x='airline',y='price',palette = "colorblind")

plt.title("Price Variability by Airline", fontsize=14, fontweight='bold')
plt.xlabel("Airline")
plt.ylabel("Standard Deviation of Price")
plt.xticks(rotation=45)

plt.show()
```

```python
# ---------- Q7. Source cities with highest duration ----------
#From which source city do flights have the highest average duration?
# Which city’s flights take longer on average?

# Group and convert to DataFrame
sour_city_highest_dur = df.groupby('source_city')['duration'].mean().reset_index()

# Sort descending
sour_city_highest_dur = sour_city_highest_dur.sort_values(by='duration', ascending=False)

print("\nDuration by Source:\n", sour_city_highest_dur)

# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(6,4))

sns.barplot(data=sour_city_highest_dur,y='source_city',x='duration',palette = "muted")

plt.title("Average Duration by Source City", fontsize=14, fontweight='bold')
plt.xlabel("Average Duration")
plt.ylabel("Source City")

plt.show()
```

```python
# ---------- Q8. Non-stop vs stops ----------
#How does the number of stops affect ticket price?
# Are non-stop flights more expensive?
# Are flights with 1 or 2 stops cheaper?
# Does adding stops reduce price?


# Group and convert to DataFrame
non_stop_vs_stops = df.groupby('stops')['price'].mean().reset_index()

print("\nStops vs Price:\n", non_stop_vs_stops)

# Set seaborn style
sns.set_style("whitegrid")

plt.figure(figsize=(7,7))

colors = sns.color_palette("Set2")  # Colorful palette

plt.pie(non_stop_vs_stops['price'],labels=non_stop_vs_stops['stops'],autopct='%1.1f%%',colors=colors,startangle=140,shadow=True)

plt.title("Average Price Share by Stops", fontsize=14, fontweight='bold')

plt.show()
```

```python
#--------------------Q9. Correlation duration & price ----------
# Calculate correlation
corr_dur_price = df['duration'].corr(df['price'])
print("\nCorrelation Duration-Price:", corr_dur_price)

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(8,6))

sns.regplot(x='duration',y='price',data=df,scatter_kws={'alpha':0.4},line_kws={'color':'red'})

plt.title("Duration vs Price", fontsize=14, fontweight='bold')
plt.xlabel("Duration")
plt.ylabel("Price")

plt.show()
```

```python
# ---------- Q10. Price by destination ----------

# Group and convert to DataFrame
price_by_dest = df.groupby('destination_city')['price'].mean().reset_index()
# Sort descending
price_by_dest = price_by_dest.sort_values(by='price', ascending=False)
print("\nDestination Price:\n", price_by_dest)

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(10,6))

sns.barplot(data=price_by_dest,x='destination_city',y='price',palette = 'pastel')

plt.title("Average Price by Destination", fontsize=16, fontweight='bold')
plt.xlabel("Destination City")
plt.ylabel("Average Ticket Price")
plt.xticks(rotation=45)

plt.show()
```

```python
# ---------- Q11. Arrival time impact ----------

# Group and convert to DataFrame
arrv_time_imp = df.groupby('arrival_time')['price'].mean().reset_index()
# Sort by price
arrv_time_imp = arrv_time_imp.sort_values(by='price')
print("\nArrival Impact:\n", arrv_time_imp)
# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(9,6))

sns.barplot(data=arrv_time_imp,x='arrival_time',y='price',palette = 'Set2')

plt.title("Arrival Time vs Average Price", fontsize=16, fontweight='bold')
plt.xlabel("Arrival Time")
plt.ylabel("Average Ticket Price")
plt.xticks(rotation=45)

plt.show()
```

```python
#---------------------Q12. Business always higher? ----------
# Are Business Class flights consistently priced higher regardless of the airline?
cls_price_compar_by_AL = df.pivot_table(values='price', index='airline', columns='class', aggfunc='mean')
print("\n Business vs Economy by Airline:\n", cls_price_compar_by_AL)


cls_price_compar_by_AL.plot(kind='bar')
plt.title("Business vs Economy Price by Airline")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
#----------------------Q13. Most business flights ----------
# “Which airline operates the highest number of Business class flights?”


# Filter Business class flights
most_business_flights = df[df['class'] == 'Business']['airline'].value_counts().reset_index()

# Rename columns properly
most_business_flights.columns = ['airline', 'count']

print("\nMost Business Flights:\n", most_business_flights)

# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(6,4))

sns.barplot(data=most_business_flights,x='airline',y='count',palette = 'Dark2')

plt.title("Business Flights Count by Airline", fontsize=16, fontweight='bold')
plt.xlabel("Airline")
plt.ylabel("Number of Business Flights")
plt.xticks(rotation=45)
plt.show()
```

```python
# ---------- Q14. Routes with price fluctuation ----------


# Calculate fluctuation and get top 10 directly
routes_price_fluct = (df.groupby(['source_city','destination_city','days_left'])['price'].mean().groupby(level=[0,1]).std()
                      .nlargest(10)          # directly get top 10
                      .reset_index(name='std_dev'))

# Create route column
routes_price_fluct['route'] = (
    routes_price_fluct['source_city'] + " → " +
    routes_price_fluct['destination_city'])

print("\nTop 10 Price Fluctuation:\n", routes_price_fluct)

# Plot
sns.set_style("whitegrid")

plt.figure(figsize=(9,6))
sns.barplot(data=routes_price_fluct,
            y='route',
            x='std_dev',
            palette='seismic')

plt.title("Top 10 Routes with Highest Price Fluctuation")
plt.xlabel("Price Standard Deviation")
plt.ylabel("Route")

plt.tight_layout()
plt.show()
```

```python
# ---------- Q15. Airlines dominating routes ----------
Dominate_AL_Routes = df.groupby(['source_city','destination_city'])['airline']\
       .agg(lambda x: x.value_counts().idxmax())
print("\n Dominating Airlines:\n", Dominate_AL_Routes.head(10))

plt.figure()
Dominate_AL_Routes.value_counts().plot(kind='bar')
plt.title("Airline Dominance Count Across Routes")
plt.ylabel("Number of Routes Dominated")
plt.show()
```
